```python
import pandas as pd
import numpy as np
%matplotlib inline
import re
```

```python
stores_df = pd.read_csv('../Data/stores.csv')
```

```python
stores_df['city'].unique()
```

```python
transactions_df = pd.read_csv('../Data/transactions.csv')
```

```python
transactions_df.head()
```

```python
transactions_df.info()
```

```python
store_transactions = transactions_df.merge(stores_df,how='inner',on='store_nbr')
```

```python
store_transactions.head()
```

```python
store_transactions.info()
```

```python
store_transactions['state'].unique()
```

```python
store_transactions['city'].unique()
```

```python
max_store_loc = store_transactions.groupby('city')['store_nbr'].max().reset_index().sort_values(by='store_nbr',ascending=True)
min_store_loc = store_transactions.groupby('city')['store_nbr'].min().reset_index().sort_values(by='store_nbr',ascending=True)

max_store_loc.merge(min_store_loc,how='inner',on='city')
```

```python
max_store_loc = store_transactions.groupby('state')['store_nbr'].max().reset_index().sort_values(by='store_nbr',ascending=True)
min_store_loc = store_transactions.groupby('state')['store_nbr'].min().reset_index().sort_values(by='store_nbr',ascending=True)

max_store_loc.merge(min_store_loc,how='inner',on='state')
```
